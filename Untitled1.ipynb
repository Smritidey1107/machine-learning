# Task
Analyze the "snu_friendship.csv" dataset to cluster students into "friendship groups" based on shared hobbies and club interests, following the steps outlined in the plan.

## Load and inspect the data

### Subtask:
Load the `snu_friendship.csv` file into a pandas DataFrame and display the first few rows and data types to understand the data structure.


**Reasoning**:
Load the data into a pandas DataFrame and display the first few rows and data types to understand the data structure.



```python
import pandas as pd

df = pd.read_csv('/content/snu_friendship.csv')
display(df.head())
display(df.info())
```

## Preprocess the data

### Subtask:
Handle categorical features using one-hot encoding, as clustering algorithms typically require numerical input. Identify the categorical columns and apply one-hot encoding to prepare the data for clustering.


**Reasoning**:
Identify and one-hot encode the categorical columns in the DataFrame.



```python
categorical_cols = df.select_dtypes(include='object').columns
df_encoded = pd.get_dummies(df, columns=categorical_cols, dummy_na=False)
df_encoded.columns = df_encoded.columns.str.replace(r'\s+', '_', regex=True).str.replace(r'[^a-zA-Z0-9_]', '', regex=True)
display(df_encoded.head())
display(df_encoded.info())
```

## Determine the optimal number of clusters

### Subtask:
Use the silhouette score and Davies–Bouldin index to evaluate the quality of clustering for different numbers of clusters (from 2 to 10) and identify the optimal number. Visualize the silhouette scores and Davies-Bouldin indices against the number of clusters.


**Reasoning**:
Evaluate the clustering quality for different numbers of clusters using silhouette score and Davies-Bouldin index and visualize the results to find the optimal number of clusters.



```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt

silhouette_scores = []
davies_bouldin_indices = []

for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    kmeans.fit(df_encoded)
    silhouette_scores.append(silhouette_score(df_encoded, kmeans.labels_))
    davies_bouldin_indices.append(davies_bouldin_score(df_encoded, kmeans.labels_))

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].plot(range(2, 11), silhouette_scores)
axes[0].set_xlabel("Number of Clusters")
axes[0].set_ylabel("Silhouette Score")
axes[0].set_title("Silhouette Score vs. Number of Clusters")

axes[1].plot(range(2, 11), davies_bouldin_indices)
axes[1].set_xlabel("Number of Clusters")
axes[1].set_ylabel("Davies-Bouldin Index")
axes[1].set_title("Davies-Bouldin Index vs. Number of Clusters")

fig.suptitle("Clustering Evaluation Metrics")
plt.tight_layout()
plt.show()
```

## Perform k-means clustering

### Subtask:
Apply the K-Means algorithm with the optimal number of clusters to group students into friendship groups. Instantiate and fit the KMeans model to the encoded data and add the cluster labels to the original dataframe.


**Reasoning**:
Apply KMeans clustering with the optimal number of clusters and add the cluster labels to the original dataframe.



```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
kmeans.fit(df_encoded)
df['Cluster Label'] = kmeans.labels_
display(df.head())
```

## Visualize the clusters

### Subtask:
Use a dimensionality reduction technique like PCA to reduce the dimensions of the data to 2 components and visualize the clusters in a 2D plot to gain insights into the grouping.


**Reasoning**:
Apply PCA to reduce the dimensions of the encoded data to 2 components, create a new DataFrame from the transformed data, and then create a scatter plot to visualize the clusters in 2D, coloring the points by their cluster label.



```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2, random_state=42)
pca_components = pca.fit_transform(df_encoded)

df_pca = pd.DataFrame(data=pca_components, columns=['PCA Component 1', 'PCA Component 2'])

plt.figure(figsize=(10, 8))
scatter = plt.scatter(df_pca['PCA Component 1'], df_pca['PCA Component 2'], c=df['Cluster Label'], cmap='viridis')
plt.title('PCA of Friendship Groups')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(scatter, label='Cluster Label')
plt.show()
```

## Evaluate the clusters

### Subtask:
Calculate and display the silhouette score and Davies–Bouldin index for the final clustering result to quantify the quality of the clusters.


**Reasoning**:
Calculate and display the silhouette score and Davies–Bouldin index for the final clustering result.



```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

silhouette = silhouette_score(df_encoded, df['Cluster Label'])
davies_bouldin = davies_bouldin_score(df_encoded, df['Cluster Label'])

print(f"Silhouette Score: {silhouette}")
print(f"Davies-Bouldin Index: {davies_bouldin}")
```

## Present the results

### Subtask:
Display the number of students in each cluster and provide insights into the characteristics of each cluster based on the original features (numerical statistics and top categorical values).


**Reasoning**:
Calculate and display the number of students in each cluster and then iterate through each cluster to calculate and print descriptive statistics for numerical columns and identify and print the top categorical values for categorical columns.



```python
print("Number of students in each cluster:")
display(df['Cluster Label'].value_counts().sort_index())

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include='object').columns.tolist()

for cluster_label in sorted(df['Cluster Label'].unique()):
    print(f"\n--- Characteristics for Cluster {cluster_label} ---")
    cluster_df = df[df['Cluster Label'] == cluster_label]

    print("\nNumerical Feature Statistics:")
    display(cluster_df[numerical_cols].describe())

    print("\nTop Categorical Values:")
    for col in categorical_cols:
        top_categories = cluster_df[col].value_counts().nlargest(3) # Get top 3 categories
        if not top_categories.empty:
            print(f"- {col}:")
            display(top_categories)
```
